```python
import numpy as np
from array import*
ds=df.iloc[:,:]
dm=df.iloc[:,19:23].values
severity=[]
for i in dm:
  a=i.tolist()
  if (a==[1,0,0,0]):
    severity.append(1)
  elif(a==[0,1,0,0]):
    severity.append(2)
  elif(a==[0,0,1,0]):
    severity.append(0)
  elif(a==[0,0,0,1]):
    severity.append(3)
  else:
    severity.append(np.nan)
ds=ds.assign(Severity=severity)

da=ds.iloc[:,11:16].values
age=[]
for i in da:
  a=list(i)
  if (a==[1, 0, 0, 0, 0]):
    age.append(0)
  elif (a==[0, 1, 0, 0, 0]):
    age.append(1)
  elif (a==[0, 0, 1, 0, 0]):
    age.append(2)
  elif (a==[0, 0, 0, 1, 0]):
     age.append(3)
  elif (a==[0, 0, 0, 0, 1]):
    age.append(4)
  else:
    age.append(np.nan)
ds=ds.assign(Age=age)

dp=ds.iloc[:,16:19].values
gender=[]
for i in dp:
  a=list(i)
  if (a==[1, 0, 0]):
    gender.append(2)
  elif (a==[0, 1, 0]):
    gender.append(1)
  elif (a==[0, 0, 1]):
    gender.append(0)
  else:
    gender.append(np.nan)
ds=ds.assign(Gender=gender)

dc=ds.iloc[:,23:26].values
contact=[]
for i in dc:
  a=list(i)
  if (a==[1, 0, 0]):
    contact.append(1)
  elif (a==[0, 1, 0]):
    contact.append(0)
  elif (a==[0, 0, 1]):
    contact.append(2)
  else:
    contact.append(np.nan)
ds=ds.assign(Contact=contact)

ds.drop(columns="Age_0-9",axis=1,inplace=True)
ds.drop(columns="Age_10-19",axis=1,inplace=True)
ds.drop(columns="Age_20-24",axis=1,inplace=True)
ds.drop(columns="Age_25-59",axis=1,inplace=True)
ds.drop(columns="Gender_Female",axis=1,inplace=True)
ds.drop(columns="Gender_Male",axis=1,inplace=True)
ds.drop(columns="Gender_Transgender",axis=1,inplace=True)
ds.drop(columns="Severity_Mild",axis=1,inplace=True)
ds.drop(columns="Severity_Moderate",axis=1,inplace=True)
ds.drop(columns="Severity_None",axis=1,inplace=True)
ds.drop(columns="Severity_Severe",axis=1,inplace=True)
ds.drop(columns="Contact_Dont-Know",axis=1,inplace=True)
ds.drop(columns="Contact_No",axis=1,inplace=True)
ds.drop(columns="Contact_Yes",axis=1,inplace=True)
ds.head()
```

```python
x=df.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
key = ['LogisticRegression','KNeighborsClassifier','SVC','DecisionTreeClassifier','RandomForestClassifier','GradientBoostingClassifier','AdaBoostClassifier','XGBClassifier']
value = [LogisticRegression(solver='lbfgs',max_iter=350000),KNeighborsClassifier(),SVC(kernel='rbf'),DecisionTreeClassifier(),RandomForestClassifier(),GradientBoostingClassifier(),AdaBoostClassifier(),xgb.XGBClassifier()]
models = dict(zip(key,value))
```

```python
for name,algo in models.items():
  model=algo
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  print(name)
  print(confusion_matrix(y_test,y_pred))
  print(accuracy_score(y_test,y_pred))
```
